```python
import numpy as np
from numpy import float64
n = 100
def xnm (n):
    xnm = np.zeros([n,n],dtype=float64)
    for i in range(1,n):
        for j in  range(1,n):
            if i == j:
                xnm[i,j] =1/2
            else:
                xnm[i,j] = ((1-(-1)**(i+j))/np.pi**2 )*((1/(i+j)**2) - (1/(i-j)**2))
    xnmf =xnm[1:,1:]
    #print(xnmf)
    return xnmf
xnm =xnm(n)
xnm.size
```
